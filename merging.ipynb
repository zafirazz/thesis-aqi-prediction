## Now that we got two csv files from each station in Debrecen, we will process it and merge it into one.

```python
import os
import pandas as pd
def merge_csv_files(input_folder: str, output_file: str):
    csv_files = [file for file in os.listdir(input_folder) if file.endswith('.csv')]
    
    if not csv_files:
        print("No CSV files found in the folder.")
        return
    
    dataframes = {}
    for i, file in enumerate(csv_files):
        df = pd.read_csv(os.path.join(input_folder, file), sep=";", decimal=",")
        df.rename(columns={col: f"Station{i+1}_{col}" for col in df.columns if col != "Date"}, inplace=True)
        dataframes[f"Station{i+1}"] = df
    
    merged_df = None
    for df in dataframes.values():
        if merged_df is None:
            merged_df = df
        else:
            merged_df = pd.merge(merged_df, df, on="Date", how="outer")
    
    merged_df.to_csv(output_file, index=False, sep=";", decimal=",")
    print(f"Merged {len(csv_files)} files into {output_file}")

if __name__ == "__main__":
    input_folder = "/Users/zafiraibraeva/Code/uni coding/thesis/dataset/full_data/daily_2025" 
    output_file = "/data/AQI data/AQI_2025.csv"
    merge_csv_files(input_folder, output_file)
```

```python
import os
import pandas as pd

def merge_csv_files(source_folder: str, destination_folder: str, output_filename: str = "wind.csv"):
    """
    Merges all CSV files in the source folder into a single CSV file and saves it in the destination folder.
    
    :param source_folder: Path to the folder containing the CSV files to merge.
    :param destination_folder: Path to the folder where the merged CSV file will be saved.
    :param output_filename: Name of the output merged CSV file (default: "wind.csv").
    """
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
    
    csv_files = [f for f in os.listdir(source_folder) if f.endswith('.csv')]
    
    if not csv_files:
        print("No CSV files found in the source folder.")
        return
    
    dataframes = [pd.read_csv(os.path.join(source_folder, file)) for file in csv_files]
    merged_df = pd.concat(dataframes, ignore_index=True)
    
    output_path = os.path.join(destination_folder, output_filename)
    merged_df.to_csv(output_path, index=False)
    
    print(f"Merged {len(csv_files)} CSV files into {output_path}")

# Example usage
source_folder = "/Users/zafiraibraeva/Code/uni coding/thesis/thesis_code/thesis/data/wind"
destination_folder = "/Users/zafiraibraeva/Code/uni coding/thesis/thesis_code/thesis/data/final_data/wind.csv"
merge_csv_files(source_folder, destination_folder)
```
